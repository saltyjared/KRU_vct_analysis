```python
def get_unique_match_stats(df):
    new_df = df[~df['Agents'].str.contains(',')]
    new_df = new_df.groupby(['Match Type', 'Player', 'Agents']).max('Rounds Played').reset_index()
    return new_df
```

```python
kru_2023_reg = get_unique_match_stats(kru_2023[kru_2023['Tournament'] == 'Champions Tour 2023: Americas League'])
kru_2023_lcq = get_unique_match_stats(kru_2023[kru_2023['Tournament'] == 'Champions Tour 2023: Americas Last Chance Qualifier'])

kru_2024_kickoff = get_unique_match_stats(kru_2024[(kru_2024['Tournament'] == 'Champions Tour 2024: Americas Kickoff')])
kru_2024_reg = get_unique_match_stats(kru_2024[(kru_2024['Tournament'] == 'Champions Tour 2024: Americas Stage 1') & (kru_2024['Stage'] == 'Regular Season')])
kru_2024_playoffs = get_unique_match_stats(kru_2024[(kru_2024['Tournament'] == 'Champions Tour 2024: Americas Stage 1') & (kru_2024['Stage'] == 'Playoffs')])
```

# KRÜ in 2023

KRÜ Esports was perhaps one of the most interesting teams in all of VCT 2023, losing every single match during the regular season, often times in heartbreaking fashion. Looking at the round differential across the Americas teams reveals a shocking revelation: KRÜ was not the worst team. That title belongs to MIBR, who still managed to pick up 3 wins during the season despite having an overall round differential of -72, compared to KRÜ's -53. Let's visualize week by week how the team did across several key metrics in Valorant player performance.

```python
import plotly.express as px
import plotly.graph_objects as go
from ipywidgets import widgets
import matplotlib.pyplot as plt

data = kru_2023_reg
data
```

```python
data.columns.tolist()
```

```python
week = widgets.SelectionSlider(
    options=data['Match Type'].unique().tolist(),
    value='All Match Types',
    disabled=False,
    continous_update=False,
    orientation='horizontal',
    readout=True
)

stat = widgets.Dropdown(
    options=data.columns.tolist()[3:],
    value='Rating',
    description='Stat:'
)

options = widgets.HBox(children=[week, stat])

players = []
for player in sorted(data['Player'].unique().tolist(), key=str.casefold):
    player_box = widgets.Checkbox(
        value=False,
        description=player,
        disabled=False,
        indent=False
    )
    players.append(player_box)

player_container = widgets.HBox(children=players)

def update(week, stat, **player_selection):
    selected_players = [player for player, selected in player_selection.items() if selected]
    if not selected_players:
        plt.figure(figsize=(10, 6))
        plt.bar([0], [0])
        plt.xlabel('No player selected')
        plt.ylabel(stat)
        plt.title(f'{stat} for Selected Players in {week}')
        ax = plt.gca()
        ax.axes.xaxis.set_ticks([])
        ax.axes.yaxis.set_ticks([])
        return
    
    filtered = data[(data['Match Type'] == week) & (data['Player'].isin(selected_players))]

    if filtered.empty:
        display("No data available for the selected options.")
        return
    
    plt.figure(figsize=(10, 6))
    plt.bar(filtered['Player'], filtered[stat])
    plt.xlabel('Player')
    plt.ylabel(stat)
    plt.title(f'{stat} for Selected Players in {week}')
    plt.show()
```

```python
interactive_output = widgets.interactive_output(update, {
    'week':week,
    'stat':stat,
    **{player.description: player for player in players}
})

display(widgets.VBox([options, player_container]))
display(interactive_output)
```
