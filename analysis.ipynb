The above cell shows the 21 total columns contained in each of the match datasets. One stat I'll be adding is `First Engagements` or the sum of `First Kills` and `First Deaths`, which is a metric often used to show which players are play the primary entry role.

```python
vct_2023_data['First Engagements'] = vct_2023_data['First Kills'] + vct_2023_data['First Deaths']
vct_2024_data['First Engagements'] = vct_2024_data['First Kills'] + vct_2024_data['First Deaths']
print('First 5 rows of the VCT 2023 match data:\n')
display(vct_2023_data.head())
print('First 5 rows of the VCT 2024 match data:\n')
display(vct_2024_data.head())
```
